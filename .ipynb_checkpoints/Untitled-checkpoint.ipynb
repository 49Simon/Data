```python
import numpy as np
from tensorflow import keras
import os
import matplotlib as mp
import matplotlib.pyplot as plt
import pandas as pd
```

read the files

```python
testdata = pd.read_csv("data.csv")
label = testdata['Connected'] 
traindata = testdata[['linkQuality','neighborLinkQuality','RSSI value', 'AVG RSSI value']]
testdata
#traindata contains the LQ,NLQ,RSSI,AVERSSI


```

Try to put them in a graph

```python
fig, ax = plt.subplots()
ax.set_xlabel("Time")
ax.plot(testdata['systemTime']-testdata['systemTime'][0],traindata['linkQuality'],label="LQ")
ax.plot(testdata['systemTime']-testdata['systemTime'][0],traindata['neighborLinkQuality'],label="NLQ")
#ax.scatter(traindata['neighborLinkQuality'],label, label="NLQ")
ax.legend()
fig, asignal = plt.subplots()
asignal.set_xlabel("Time")
asignal.plot(testdata['systemTime']-testdata['systemTime'][0],traindata['RSSI value'],label="RSSI")
asignal.plot(testdata['systemTime']-testdata['systemTime'][0],traindata['AVG RSSI value'],label="AVG RSSI")
asignal.legend()

```
